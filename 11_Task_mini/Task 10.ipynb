***Detailed summary report***

The first table will be a summary of the distance (km) and climb (m) variables for each training activity. The second table will list the summary statistics for the average speed (km/hr), climb (m), and distance (km) variables for each training activity.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_activities = pd.read_csv("datasets\cardioActivities.csv",parse_dates = True , index_col= "Date")
df_activities
#df_activities.tail(40)
```

```python
df_walk = df_activities[df_activities["Type"]=="Walking"].copy()
print(df_walk.head(5))
df_walk["Average Heart Rate (bpm)"].fillna(110,inplace=True)
df_walk.drop(["Route Name","Friend's Tagged"], inplace=True, axis=1)
print(df_walk.isnull().sum())
```

```python
df_cycle = df_activities[df_activities["Type"]=="Cycling"].copy()
df_cycle["Average Heart Rate (bpm)"].fillna(124,inplace=True)
print(df_cycle.isnull().sum())
```

```python
df_run_walk_cycle = df_activities.append([df_walk, df_cycle]).sort_index(ascending=False)
df_run_walk_cycle.head(5)
```

```python
dist_climb_cols, speed_col = ["Distance (km)","Climb (m)"],["Average Heart Rate (bpm)"]
```

```python
df_totals = df_run_walk_cycle.groupby("Type")[dist_climb_cols].sum()
display(df_totals)
```

```python
df_summary = df_run_walk_cycle.groupby("Type")[dist_climb_cols + speed_col].describe()
df_summary
```

```python
for i in dist_climb_cols:
  df_summary[i, "total"] = df_totals[i]
print("Total Statistics")
df_summary.stack()
```
