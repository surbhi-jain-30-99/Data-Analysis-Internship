***Visualization with averages***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_activities = pd.read_csv("datasets\cardioActivities.csv",parse_dates = True , index_col= "Date")
df_activities
```

```python
df_run=df_activities.loc[df_activities['Type'] == 'Running'].copy()
runs_subset_2015_2018 = df_run.loc["2018-11-11 14:05:12":"2015-01-02 18:35:11"] 
runs_subset_2015_2018
```

```python
runs_distance = runs_subset_2015_2018["Distance (km)"]
runs_hr       = runs_subset_2015_2018["Average Heart Rate (bpm)"]
print(runs_distance)
print("-------------------------------------------------------------------------------------------------------------------------")
print(runs_hr)
```

sharex:Controls sharing of properties among x ( sharex ) axis: If sharex is set to True or all , the x-axis will be shared among all subplots. If sharex is set to False or none , each x-axis of a subplot will be independent. If it is set to row , each subplot row will share an x-axis.


axhline : axhline() function in axes module of matplotlib library is used to add a horizontal line across the axis.


```python
fig, (ax1,ax2) = plt.subplots(2, sharex=True, figsize=(12,8))
runs_distance.plot(ax=ax1 , color="red")  #first plot
ax1.set(ylabel="Distance (km)", title="Avergae Data")
ax1.axhline(runs_distance.mean(),color="black",linewidth = 1, linestyle = '-.')

runs_hr.plot(ax=ax2, color="black")
ax2.set(xlabel="Date", ylabel="Average Heart Rate (bpm)")
ax2.axhline(runs_hr.mean(), color="black",linewidth = 1, linestyle = '-.')

plt.show()
```
