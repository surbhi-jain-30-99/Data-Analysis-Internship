```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_activities = pd.read_csv("datasets\cardioActivities.csv",parse_dates = True , index_col= "Date")
df_activities
df_activities.tail(40)
```

```python
df_run=df_activities.loc[df_activities['Type'] == 'Running'].copy()
df_run
```

```python
runs_subset_2013_2018 = df_run.loc["2018-11-11 14:05:12":"2013-01-02 18:35:11"] 
runs_subset_2013_2018
```

```python
df_run_dist_annual =  runs_subset_2013_2018["Distance (km)"].resample("A").sum()
df_run_dist_annual.head(5)
```

```python
fig = plt.figure(figsize = (8,5))
ax = df_run_dist_annual.plot(marker="*", markersize=14, linewidth=0, color="Black")
ax.set(ylim = [0,800],xlim= ["2012","2019"],xlabel = "Years",ylabel = "Distance(km)",title = "Annual distance")
ax.axhspan(0,800,color="red",alpha = 0.2)
plt.show()
```
