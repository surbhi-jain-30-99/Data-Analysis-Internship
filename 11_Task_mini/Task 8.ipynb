```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_activities = pd.read_csv("datasets\cardioActivities.csv",parse_dates = True , index_col= "Date")
df_activities
#df_activities.tail(40)
```

```python
import statsmodels.api as sm
runs_subset_2013_2018 = df_activities.loc["2018-11-11 14:05:12":"2013-01-02 18:35:11"] 
runs_subset_2013_2018
```

```python
df_run_dist_wkly =  runs_subset_2013_2018["Distance (km)"].resample("W").bfill() 
#fill NaN values with the bfill() method. 
df_run_dist_wkly.head(5)
```

```python
fig = plt.figure(figsize = (12,5))
ax = df_run_dist_wkly.plot(marker=".",label = "Trend", color="Black")
ax.legend()
ax.set_title = 'Running Title'
plt.show()
```
