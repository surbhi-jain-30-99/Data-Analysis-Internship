```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_activities = pd.read_csv("datasets\cardioActivities.csv",parse_dates = True , index_col= "Date")
df_activities
#df_activities.tail(40)
```

```python
df_run=df_activities.loc[df_activities['Type'] == 'Running'].copy()
df_run
```

```python
runs_subset_2015_2018 = df_run.loc["2018-11-11 14:05:12":"2015-01-02 18:35:11"] 
runs_subset_2015_2018
```

```python
df_run_hr_all = runs_subset_2015_2018["Average Heart Rate (bpm)"]
df_run_hr_all
```

```python
hr_zones = [90,110,130,142,150,170]
zone_names = ["low", "min", "moderate", "medium", "max","Extreme"]
fig, ax = plt.subplots(figsize=(8,5))
ax.grid(True)
ax.hist(df_run_hr_all,bins=hr_zones,color='blue')
ax.xaxis.set(ticks=hr_zones)
ax.set(title="HR DISTRIBUTIONS", ylabel="Number of runs")
ax.set_xticklabels(labels=zone_names, rotation=-30, ha="left")
plt.show()
```
