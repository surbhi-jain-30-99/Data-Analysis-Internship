```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_activities = pd.read_csv("datasets\cardioActivities.csv",parse_dates = True , index_col= "Date")
df_activities
```

```python
df_run=df_activities.loc[df_activities['Type'] == 'Running'].copy()
df_run
```

```python
df_run.head(5)
```

```python
runs_subset_2013_2018 = df_run.loc["2018-11-11 14:05:12":"2013-01-02 18:35:11"] #subset 
#.loc is label-based which means that you have to specify rows and columns based on their row and column labels
runs_subset_2013_2018.head(5)
```

```python
runs_subset_2013_2018.plot(subplots=True,figsize=(10,5), marker=".", markersize=5)
plt.show()
```
