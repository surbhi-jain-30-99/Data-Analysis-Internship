```python
x = Twelve_rows['DATE']
y = Twelve_rows['VALUE']
```

```python
import seaborn as sns
plt.figure(figsize=(10,10))
sns.lineplot(x,y)
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```
